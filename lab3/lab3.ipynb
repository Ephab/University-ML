```python
import pandas as pd
import numpy as np
```

```python
import kagglehub
path = kagglehub.dataset_download("alexteboul/diabetes-health-indicators-dataset")
```

```python
import pathlib
path = pathlib.Path(path)
csv = next(path.iterdir())
```

```python
df = pd.read_csv(csv)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()
```

```python
income_copy = df['Income'].copy()
df.Income = df.Income.map({
    1: "18-24",
    2: "25-29",
    3: "30-34",
    4: "35-39",
    5: "40-44",
    6: "45-49",
    7: "50-54",
    8: "55-59",
    9: "60-64",
    10: "65-69",
    11: "70-74",
    12: "75-79",
    13: "80+"
})

counts = df['Income'].value_counts()
```

```python
plt.figure(figsize=(8, 5))
plt.bar(counts.index, height=counts)
plt.xlabel("Income Level")
plt.ylabel("Count")
plt.title("Distribution of Income Levels")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df["MentHlth"], bins=31, kde=True)
plt.title("Distribution of Mental Health Days (Past 30 Days)", fontsize=14)
plt.xlabel("Number of Days with Poor Mental Health")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
ax = sns.countplot(data=df, x="HighBP")
plt.title("Distribution of High Blood Pressure")
plt.xlabel("High Blood Pressure (0 = No, 1 = Yes)")
plt.ylabel("Count")
```

```python
cols = ["BMI", "GenHlth", "MentHlth", "PhysHlth", "Age", "HighBP", "HighChol"]

corr = df[cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(
    corr,
    annot=True,
)
plt.title("Heatmap of Health Features")
plt.yticks(rotation=0)
plt.show()
```

```python
age_diabetes = df.groupby('Age')['Diabetes_012'].mean().sort_index()

plt.figure(figsize=(10, 5))
age_diabetes.plot(kind='bar')

plt.title("Average Diabetes Rate by Age")
plt.xlabel("Age")
plt.ylabel("Average Diabetes Rate")
plt.xticks(rotation=0)
plt.show()
```

```python
income_diabetes = df.groupby('Income')['Diabetes_012'].mean().sort_index()

plt.figure(figsize=(10, 5))
income_diabetes.plot(kind='bar')

plt.title("Average Diabetes Rate by Income Level")
plt.xlabel("Income Level")
plt.ylabel("Average Diabetes Rate")
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='BMI', y='GenHlth', data=df, alpha=0.3)
plt.title("BMI vs General Health")
plt.xlabel("BMI")
plt.ylabel("General Health (1=Excellent, 5=Poor)")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
bp_diabetes = df.groupby('HighBP')['Diabetes_012'].mean()
bp_diabetes.plot(kind='bar')

plt.title("Average Diabetes Rate by High Blood Pressure")
plt.xlabel("High Blood Pressure (0=No, 1=Yes)")
plt.ylabel("Average Diabetes Rate")
plt.xticks(rotation=0)
plt.show()
```
